```python
import csv, json
import os
import pandas as pd

#Get the list of downloaded stock symbols in the folder
dlist1 = os.listdir('D:/StockData/HistoryData/')

for sid in dlist1:
    #Get the list of downloaded date in the stock symbol folder
    dlist2 = os.listdir('D:/StockData/HistoryData/'+sid+'/')

    #Clean and arrange field and merge files
    tempData = []
    for d in dlist2:
        try:
            op = open('D:/StockData/HistoryData/'+sid+'/'+d, 'r', encoding='big5')
            rows = csv.reader(op)
            tempData1 = []
            for row in rows:
                if row == []:
                    continue
                elif row == ['', '']:
                    continue
                tempData1.append(row)
            op.close()
            
            if not tempData1[0] == ['日期']:
                del tempData1[0]
            del tempData1[((len(tempData1)//9)*9):len(tempData1)]
            tempData2 = [[] for n in range(int(len(tempData1)/9))]
            for n in range(len(tempData1)):
                nt = int(n/9)
                tempData2[nt].append(tempData1[n][0])
            if not len(tempData) == 0:
                del tempData2[0]
            for data in tempData2:
                tempData.append(data)
        except:
            print(sid, d, 'error')
    
    #Restore the cleaned up data back
    op = open('D:/StockData/ForAnalysis/'+sid+'.csv', 'w', newline='', encoding='utf-8-sig')
    writer = csv.writer(op)
    for row in tempData:
        writer.writerow(row)
    op.close()
    print(sid+'done')
    
    
```

```python
from datetime import datetime
import pandas as pd
import os
import csv, json

#Get the list of files in the folder
dlist = os.listdir('D:/StockData/ForAnalysis/')

#Convert date format and text to number
for dn in dlist:
    op = open('D:/StockData/ForAnalysis/'+dn, 'r', encoding='utf-8')
    rows = csv.reader(op)
    tempData = []
    n = 0
    for row in rows:
        if n == 0:
            row[0] = '日期'
            tempData.append(row[0:7])
            n += 1
        else:
            try:
                row[0] = str(int(row[0][0:3])+1911)+row[0][3:]
                row[0] = datetime.strptime(row[0], '%Y/%m/%d').date()
                for n in [1, 2, 8]:
                    row[n] = int(row[n].replace(',', ''))
                for n in range(3, 7):
                    row[n] = float(row[n])
                tempData.append(row[0:7])
            except:
                print(dn, row)
                continue
    
    #Restore data
    op.close()
    op = open('D:/StockData/ForAnalysis/'+dn, 'w', newline='', encoding='utf-8-sig')
    writer = csv.writer(op)
    for row in tempData:
        writer.writerow(row)
    op.close()
    print(dn, 'done')
    
print('all done')
```
